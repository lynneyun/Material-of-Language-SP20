```python
from fontTools.ttLib import TTFont
from fontTools.pens.recordingPen import RecordingPen
from numpy import interp

font1 = TTFont("./NotoSans-Regular.ttf")
font2 = TTFont("./KevinTestfont-Regular.otf")

def glyphcommands(f, ch):
    glyph = f.getGlyphSet()[ch]
    p = RecordingPen()
    glyph.draw(p)
    return p.value

def advancefor(f, ch):
    glyph = f.getGlyphSet()[ch]
    return glyph.width

def getpath(font, ch, scale,trans_x,x_offset, y_offset,frame_num):
    rand_x_shift = interp(draw.cos(frame_num),[-1,1],[0.0,300.0])
    rand_y_shift = interp(draw.sin(frame_num),[-1,1],[0.0,300.0])
    path_orig = glyphcommands(font, ch)
    paths = []
    for i in path_orig:
        if i[0] == 'moveTo':
            path = draw.BezierPath()
            x = i[1][0][0] + rand_x_shift
            y = i[1][0][1] + rand_y_shift
            path.moveTo((x,y))
#             rand_y_shift += 10
        if i[0] == 'lineTo':
            x = i[1][0][0]+ rand_x_shift
            y = i[1][0][1]+rand_y_shift
            path.lineTo((x,y))
#             rand_y_shift += 10
        if i[0] == 'qCurveTo':
            temp = []
            for cord in i[1]:
                x = cord[0]+ rand_x_shift
                y = cord[1]+ rand_y_shift
                temp.append((x,y))
            path.qCurveTo(*temp)
#             rand_y_shift += 10
        if i[0] == 'closePath':
            path.closePath()
            paths.append(path)
#             rand_y_shift += 10
    if len(paths) > 1:
        finalpath = paths[0].difference(paths[1])
    else: 
        finalpath = paths[0]
    finalpath.scale(scale)
    finalpath.translate(trans_x + x_offset,y_offset)
    return finalpath

def combine_path(font, s,scale,x_offset,y_offset,frame_num): 
    text_paths = []
    cx = 0
    for ch in s:
        glyph_path = getpath(font, ch, scale,cx,x_offset,y_offset,frame_num)
        text_paths.append(glyph_path)
        cx += advancefor(font, ch)*scale
    combined = draw.BezierPath()
    for i in text_paths:
        combined.appendPath(i)
    return combined

###############
""

# qf_path_cmds = []
```
